```python
print(DataIris['species'].value_counts())
print()

for x in range (0, 4):
    print("Média Coluna" ,DataIris.columns[x] , ":")
    print(DataIris[DataIris.columns[x]].mean())
    print()
print()
for x in range (0, 4):
    print("Mediana Coluna" ,DataIris.columns[x] , ":")
    print(DataIris[DataIris.columns[x]].median())
    print()
```

```python
import numpy
Agrupado = DataIris.groupby('species')
Agrupado.agg([numpy.median, numpy.mean])
```

```python
import matplotlib.pyplot as MatPlot
%matplotlib inline
MatPlot.scatter(DataIris['sepal_length'], DataIris['sepal_width'])
MatPlot.xlabel('Sepal Length')
MatPlot.ylabel('Sepal Width')
MatPlot.title('Sepal Length vs Width')
```
